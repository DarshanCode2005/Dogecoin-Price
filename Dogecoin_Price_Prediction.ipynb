```python
# Import the libraries
import numpy as np
import pandas as pd
```

```python
df = pd.read_csv('DOGE-USD.csv')
```

```python
df.tail(150)
```

```python
# Set the index to be date
df = df.set_index(pd.DatetimeIndex(df['Date'].values))
```

```python
df
```

```python
#Get the close price

df = df[['Close']]
```

```python
df
```

```python
# Create a variable to store the number of days into the future that we want to predict
prediction_days = 1
# Create a new column called prediction
df['Prediction'] = df[['Close']].shift(-1*prediction_days)
```

```python
df

# so by using the above method you can also create a new column called Prediction
```

```python
# Create the independent data set x
X = np.array(df.drop(['Prediction'],axis=1))
```

```python
X
```

```python
# To actually validate my forecasting model I will purposely remove some rows
# Remove the last n+1 rows of data where n is the prediction_days

X = X[:len(df) - prediction_days - 1]
print(X[X == np.NaN])
```

```python
# Create a dependent data set (y)

y = np.array(df['Prediction'])
# Get all of the y values except the last n+1 rows
y = y[:-prediction_days - 1]

print(y[y == np.NaN])
```

```python
# Split the data into a training dataset and a testing dataset(80-20)
from sklearn.model_selection import train_test_split
x_train, x_test, y_train, y_test = train_test_split(X, y, test_size=0.2)
```

```python
nan_indices = np.isnan(y_train)
nan_indices_test = np.isnan(y_test)
lst = list(map(lambda elem: elem == True,nan_indices))
print(lst)
```

```python
y_train_cleaned = y_train[~nan_indices]
x_train_cleaned = x_train[~nan_indices]
y_test_cleaned = y_test[~nan_indices_test]
x_test_cleaned = x_test[~nan_indices_test]
```

```python
len(y_train_cleaned)
```

```python
# Use the RandomForestRegressor for the model
from sklearn.ensemble import RandomForestRegressor

forest = RandomForestRegressor(n_estimators=1,random_state=800)
forest.fit(x_train_cleaned,y_train_cleaned)

print(forest.score(x_test_cleaned,y_test_cleaned))
```

```python
# show how close the predicted and the actual values are
prediction = np.array(forest.predict(x_test))
# Print the predicted test values

# print(prediction)
print('---------------------')
# print(y_test)

np.mean(y_test_cleaned-prediction)
```
